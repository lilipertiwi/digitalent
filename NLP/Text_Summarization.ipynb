# Text Summarization
merangkum source text menjadi versi yang lebih ringkas (summary)<br>
• pahami teks<br>
• menandai bagian penting (mis. pokok kalimat di awal atau di akhir)<br>
• menemukan kata kunci<br>

**single doc :** satu dokumen<br>
**multi doc:** mungkin terjadi ovelaps<br>
**generic :** hasilnya akan sama jika dikerjakan siapa saja<br>
**domain specific :** hasil ringkasan akan berbeda tergantung yang meringkas<br>
**query based :** ada yang ditekankan di awal mengenani apa yang ingin diringkas<br>
**extractive :** hanya copy-paste bagian-bagian terpenting<br>
**abstractive :** membuat ringkasan dengan bahasa sendiri<br>
