#### Read wine.csv

```python
# Pandas merupakan salah satu library Python yang mudah digunakan untuk manipulasi data
#
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Membaca file wine.csv menggunakan method .read_csv() dari pandas
# Kemudian menyimpannya dalam variabel wine
wine = pd.read_csv('wine.csv')
wine
```

#### Describe the data

```python
# Method .describe() berfungsi untuk menampilkan informasi statistik dari kolom numerik pada dataframe wine
# Dari output dapat dilihat jumlah baris pada dataframe wine sebanyak 150930 baris. 
# Selain itu method ini menunjukkan nilai rata-rata setiap kolom, nilai standar deviasi, nilai minimal hingga maksimal
wine.describe()
```

#### Is there null values?

```python
# Menunjukkan cell mana saja dari dataframe wine yang berisi null/missing value
# Missing value ditandai dengan True
pd.isnull(wine)
```

```python
# Method .isnull di bawah mengambil semua missing value pada kolom region_1
# Kemudian dijumlahkan menggunakan method .sum()
pd.isnull(wine.region_1).sum()
```

```python
# Menunjukkan missing value yang terdapat pada kolom country
wine[pd.isnull(wine.country)]
```

#### Is there null values in region_1, region_2, points, country? Which rows?


```python
# wine[wine['region_1'].isnull()] hanya menampilkan dataframe wine di mana kolom region_1 bernilai NaN
# [['region_1']] hanya menampilkan kolom region_1
reg_1 = wine[wine['region_1'].isnull()][['region_1']]
reg_1
```

```python
# wine[wine['region_2'].isnull()] hanya menampilkan dataframe wine di mana kolom region_2 bernilai NaN
# [['region_2']] hanya menampilkan kolom region_2
reg_2 = wine[wine['region_2'].isnull()][['region_2']]
reg_2
```

```python
# wine[wine['points'].isnull()] hanya menampilkan dataframe wine di mana kolom points bernilai NaN
# [['points']] hanya menampilkan kolom points
# Output menunjukkan tidak ada data dengan kolom points bernilai NaN
point = wine[wine['points'].isnull()][['points']]
point
```

```python
# wine[wine['country'].isnull()] hanya menampilkan dataframe wine di mana kolom country bernilai NaN
# [['country']] hanya menampilkan kolom country
country = wine[wine['country'].isnull()][['country']]
country
```

```python
# Kode berikut menampilkan seluruh baris baik dari kolom region_1 atau region_2 atau points atau country yang bernilai NaN
rows = wine[wine['region_1'].isnull()|wine['region_2'].isnull()|wine['points'].isnull()|wine['country'].isnull()][['region_1','region_2','points','country']]
rows
```

```python
# Menampilkan seluruh dataframe wine yang tidak memiliki null value pada kolom country
wine[wine['country'].notnull()]
```
