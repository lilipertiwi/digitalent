```python
#9. Tampilkan tanggal, bulan dan delay kedatangan yang pernah dialami pesawat nomor 1545, terurut dari delay terlama
f1545 = flights[(flights['flight']==1545) & (flights['arr_delay'] > 0) ]
```

```python
#enlargest hanya menampilkan beberapa nilai terbesar
f1545.nlargest(10,columns='arr_delay')[['flight','day','month','arr_delay']]
```

```python
#untuk menampilkan keseluruhan, gunakan .sort_values
f1545[['flight','day','month','arr_delay']].sort_values(by='arr_delay', ascending=False)
```

```python
#10. Tunjukkan asal dan tujuan pesawat dengan carrier selain 'US','UA','DL','B6'
flights[(flights['carrier'] != 'US') & (flights['carrier'] != 'UA') & (flights['carrier'] != 'DL') & (flights['carrier'] != 'B6')][['carrier','origin','dest']]
```

```python
#atau menggunakan .isin() dan ~(not)
flights[~flights['carrier'].isin(['US','UA','DL','B6'])][['carrier','origin','dest']]
```

```python
#11. Berapa banyaknya penerbangan yang terjadi setiap bulan?
flights.groupby('month')[['flight']].count()
```

```python
flights.groupby('month').size()
```

```python
#12. Berapa banyaknya penerbangan yang terjadi setiap bulan untuk setiap carrier?
flights.groupby(['month','carrier'])[['flight']].count()
```

```python
flights.groupby(['month','carrier']).size()
```

```python
#13. Tampilkan banyaknya penerbangan yang terjadi setiap bulan untuk setiap carrier yang mencapai lebih dari 2000
flight_permonth = flights.groupby(['month','carrier'])[['flight']].count()
```

```python
flight_permonth[flight_permonth['flight'] > 2000]
```

```python
# filter menggunakan lambda 
flights.groupby(['month','carrier']).filter(lambda g: len(g)>2000).groupby(['month','carrier']).size()
```
