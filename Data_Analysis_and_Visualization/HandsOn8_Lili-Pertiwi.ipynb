#### Soal: Bagilah data Wines yang kalian dapat saat materi pertemuan-15 menjadi 60:40. Lalu Visualisasikan  data tersebut dengan scatter dengan parameter points dan price.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
wine = pd.read_csv('wine.csv')
wine.head()
```

```python
wine.shape[0]
```

```python
# method .sample mengambil sebagian data sebanyak (parameter) baris secara random dari dataframe
# parameter frac = .60 mengembalikan value float dari 60% * jumlah baris data
# parameter random_state berfungsi untuk mengembalikan nilai random yang digenerate sesuai pseudorandom number yang diinput
# sehingga random dataframe yang dihasilkan tidak akan berubah-ubah
wine_60 = wine.sample(frac = .60, random_state=20)
wine_60
```

```python
wine_60.shape[0]
```

```python
# scatter diagram menggunakan seaborn untuk kolom points dan price dari train data wine_60
plt.figure(figsize=(20,10)) # mengatur ukuran figure
sns.scatterplot(wine_60.points, wine_60.price)
```

```python
# mengambil 40% sisa dataframe setelah menghapus dataframe wine_60, sehingga menghasilkan data yang tidak saling tumpang tindih
wine_40 = wine.drop(wine_60.index)
```

```python
wine_40.shape[0]
```

```python
wine_40
```

```python
# scatter plot dari dataframe wine_40
plt.figure(figsize=(20,10)) #sns.set(rc={'figure.figsize':(20,10)})
sns.scatterplot(wine_40.points, wine_40.price)
```

```python
# regression plot untuk mempertegas garis yang dihasilkan oleh scatter plot
# berikut regression plot untuk dataframe wine_60
sns.regplot(x='points', y='price', data=wine_60)
```

```python
# contoh train data menggunakan dataframe yang sudah dibersihkan (tidak ada missing value)
wine_clean = wine.dropna()
wine_clean.shape
```

```python
wine_train = wine_clean.sample(frac=.60, random_state=20)
wine_train.shape[0]
```

```python
sns.scatterplot(wine_train.points, wine_train.price)
```

```python
wine_test = wine_clean.drop(wine_train.index)
wine_test.shape[0]
```

```python
sns.scatterplot(wine_test.points, wine_test.price)
```
